```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Hotel_Reviews.csv')
```

```python
data['Review']=data['Negative_Review']+data['Positive_Review']
data['is_bad_review']=data['Reviewer_Score'].apply(lambda x:1 if x<5 else 0)
```

```python
data['Review']=data['Review'].apply(lambda x: x.replace('No Negative','').replace('No Positive',''))
```

```python
Review=data[['Hotel_Name','Reviewer_Score','Review','is_bad_review']]
```

```python
Review = Review.sample(frac = 0.1, replace = False, random_state=42)
```

```python
Review.head()
```

```python
#Remove Punctuation
import string
string.punctuation
```

```python
def remove_punc(text):
    no_punc_text=''.join([char for char in text if char not in string.punctuation])
    return no_punc_text
```

```python
Review['no_punc_text']=Review['Review'].apply(lambda x: remove_punc(x))
```

```python
Review.head()
```

```python
#Tokenization
import re
```

```python
def tokenize(text):
    tokens=re.split('\W+',text)
    return tokens
```

```python
Review['text_tokenized']=Review['no_punc_text'].apply(lambda x: tokenize(x))
```

```python
Review.head()
```

```python
#Remove Stopwords
import nltk
from nltk.corpus import stopwords
#nltk.download('stopwords')
stopword=nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(text):
    no_stopwords=[word for word in text if word not in stopword]
    return no_stopwords
```

```python
Review['no_stopwords']=Review['text_tokenized'].apply(lambda x: remove_stopwords(x))
```

```python
Review.head()
```

```python
#Remove empty tokens
def no_empty_tokens(text):
    tokens=[char for char in text if len(char)>0]
    return tokens
Review['no_stopwords']=Review['no_stopwords'].apply(lambda x:no_empty_tokens(x))
```

```python
Review.head()
```

```python
#lemmatize
nltk.download('wordnet')
```

```python
wn=nltk.WordNetLemmatizer()
```

```python
print(wn.lemmatize('sold'))
```

```python
def lemmatizing(text):
    token_text=[wn.lemmatize(word) for word in text]
    return token_text
```

```python
Review['lemmatized_text']=Review['no_stopwords'].apply(lambda x: lemmatizing(x))
```

```python
Review['clean_data']=Review['lemmatized_text'].apply(lambda x:' '.join(x))    
```

```python
Review.head()
```

```python
#frequently used words by reviewers
from wordcloud import WordCloud
```

```python
def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 30, 
        scale = 3,
        random_state = 42
    ).generate(str(data))

    fig = plt.figure(1, figsize = (15, 15))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)

    plt.imshow(wordcloud)
    plt.show()
    
# print wordcloud
#positive review
show_wordcloud(Review[Review['is_bad_review']==0]['lemmatized_text'],title='Positive Review')
#negative review
show_wordcloud(Review[Review['is_bad_review']==1]['lemmatized_text'],title='Negative Review')
#review
show_wordcloud(Review['lemmatized_text'],title='Total Review')
```

```python
nltk.download('vader_lexicon')
```

```python
#sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
Review["sentiments"] = Review["Review"].apply(lambda x: sid.polarity_scores(x))
Review = pd.concat([Review.drop(['sentiments'], axis=1), Review['sentiments'].apply(pd.Series)], axis=1)
```

```python
sid.polarity_scores('i had a good breakfast')
```

```python
#no of characters
Review['# of char']= Review['Review'].apply(lambda x: len(x))
```

```python
#no of words
Review['# of words']=Review['Review'].apply(lambda x:len(x.split()))
```

```python
# doc to vec columns
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(Review['clean_data'].apply(lambda x: x.split(" ")))]

# train a Doc2Vec model with our text data
model = Doc2Vec(documents, vector_size=5, window=2, min_count=1, workers=4)

# transform each document into a vector data
doc2vec_df = Review['clean_data'].apply(lambda x: model.infer_vector(x.split(" "))).apply(pd.Series)
doc2vec_df.columns = ['doc2vec_vector_' + str(x) for x in doc2vec_df.columns]
Review = pd.concat([Review, doc2vec_df], axis=1)
```

```python
Review.head()
```

```python
#vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
```

```python
tfidf_result=tfidf.fit_transform(Review['clean_data']).toarray()
```

```python
tfidf_df=pd.DataFrame(tfidf_result,columns=tfidf.get_feature_names())
```

```python
tfidf_df.columns = ["word_" + str(x) for x in tfidf_df.columns]
```

```python
tfidf_df.index = Review.index
Review = pd.concat([Review, tfidf_df], axis=1)
```

```python
Review[Review['is_bad_review']==0]['compound'].plot(kind='density')
Review[Review['is_bad_review']==1]['compound'].plot(kind='density')
plt.show()
```

```python
features=[c for c in Review.columns if c not in ['Hotel_Name','Review','no_punc_text','text_tokenized','no_stopwords','lemmatized_text','clean_data','is_bad_review']]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=Review[features]
y=Review['is_bad_review']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train,y_train)
```

```python
# feature importance
feature_importances_df = pd.DataFrame({"feature": features, "importance": rf.feature_importances_}).sort_values("importance", ascending = False)
feature_importances_df.head(20)
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print('confusion matrix :',confusion_matrix(y_test,y_pred))
print('classification report :',classification_report(y_test,y_pred))
```

```python
# As the data is unbalanced the reports might not be correct
# ROC curve
from sklearn.metrics import roc_curve,auc,roc_auc_score

# predict probabilities
yhat = rf.predict_proba(X_test)

# retrieve the probabilities for the positive class
yhat_positive = yhat[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, yhat_positive)
```

```python
# plot no skill roc curve - the diagonal line
plt.plot([0, 1], [0, 1], linestyle='--', label='No Skill')

# plot roc curve
plt.plot(fpr, tpr, marker='.', label='Random Forest')

# show the legend
plt.legend()
plt.show()

# calculate and print AUROC
roc_auc = roc_auc_score(y_test, yhat_positive)
print('AUROC: %.3f' % roc_auc)
```
